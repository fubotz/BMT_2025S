```python
#import models and tokenizer
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
#list lang: https://github.com/facebookresearch/flores/tree/main/flores200
tokenizer = AutoTokenizer.from_pretrained("facebook/nllb-200-distilled-1.3B", src_lang="eng_Latn") #path of your directory with the finetuned

model = AutoModelForSeq2SeqLM.from_pretrained("facebook/nllb-200-distilled-1.3B", device_map={"":0})
```

```python
#upload source file and read
import codecs
file_name = "en.txt"
mt_output = []
translation_file = open("nllb.de.txt", 'w', encoding='utf-8')
target_id = "deu_Latn"

with codecs.open(file_name, 'r', 'utf-8') as src:
    for line in src:
        line = line.strip()
        encoded = tokenizer(line, return_tensors="pt", padding=True).to("cuda")              #tokenize
        generated_tokens = model.generate(**encoded, forced_bos_token_id=tokenizer.convert_tokens_to_ids(target_id))
        translation = tokenizer.batch_decode(generated_tokens, skip_special_tokens=True) #de tokenize
        print(translation[0], file=translation_file)
translation_file.close()
```
